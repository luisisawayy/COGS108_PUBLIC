## Dataset #1 : Alcohol Consumption US By State From 1985-2016, Per Capita Consumption (Litres) 

```python
'''Dataset Purpose: The purpose of this first dataset is the consumption of alcohol by state from 1985-2016
The consumption is measured in per capita consumption in litres. The variables consist of the per capita
consumption of different types of alcoholic beverages from 1985-2016, and also one considering all alcoholic
beverages combined. These variables are NumPy float64 variables. We also have variables for the US State,
US State abbreviation, and the year which we are considering, which are categorical and discrete quantitative variables.
In terms of cleaning, wrangling, and tidying the data, we will need to ensure that the dataset has no NaN values
, no redundant variables, remove any observations that are not for a particular state or not in the correct time frame
and reorganize the dataset, by US State and Year for easier EDA work in the future. The original dataset has 2040 observations
because this dataset has each state in a all years from 1977-2016 is an observation, including District Of Columbia. Hence, we get
51 * 40 = 2040'''

#Read File: 
alcohol_df = pd.read_csv('Alcohol_Consumption_US.csv')

#Rename Columns for simpler reading 
alcohol_df = alcohol_df.rename(columns = {'Beer (Per capita consumption)':'Beer (Per Capita Consumption In Litres)', 'Wine (Per capita consumption)':'Wine (Per Capita Consumption In Litres)', 'Spirits (Per capita consumption)':'Spirits (Per Capita Consumption In Litres)', 'All beverages (Per capita consumption)':'All Beverages (Per Capita Consumption In Litres)', 'State':'U.S State'})

#Delete unneeded columns (we do not need state abbreviations)  
alcohol_df = alcohol_df.drop('State_abbrev', axis = 1)

#Delete unneeded rows (we will exclusively be modeling the time frame 1985-2016) 
alcohol_df = alcohol_df[(alcohol_df['Year'] >= 1985) & (alcohol_df['Year'] <= 2016)]

#Delete the District of Columbia
alcohol_df = alcohol_df[(alcohol_df['U.S State'] != 'District of Columbia')]

#Sort the years for each state 
alcohol_df = alcohol_df.sort_values(by = ['U.S State', 'Year'])
alcohol_df = alcohol_df.reset_index()
alcohol_df = alcohol_df.drop('index', axis = 1)

#Assert that we have no nan values 
null_rows = len(alcohol_df) - len(alcohol_df.dropna()) 
assert null_rows == 0

#Assert that the years in our dataset make sense with our hypothesis 
max_year = max(alcohol_df['Year'])
min_year = min(alcohol_df['Year'])
assert max_year == 2016
assert min_year == 1985

#Assert Shape In Correct
assert alcohol_df.shape == (1600, 6)

#Print the first few values from our dataset 
alcohol_df.head()
```

## Dataset #2-#52 : Homicide Rates Per 100,000 Residents For Each State From 1985-2016. The Datasets For Each State Will Be Combined Into One

```python
'''Dataset Purpose: The purpose of the 50 datasets is to give us information about the homicide rates per 100,000 residents in each state from
1985-2016. We mentioned above that each of the 50 datasets has 385 variables and 4 observations. 384 variables comes from the fact that the 
datasets for each state has the homicide rates per month per 100,000 residents. Hence, for 12 months, and 32 years from 1985-2016, we have a
total of 384 variables. Each month and year is a variable, or a column in the dataset. In each month and year, we have the homicide rate per 100,000
which is a object datatype which will get converted to a numeric datatype in EDA. The last variable is the "series" variable which essentially categorizes the data by location and type of metric. For example, 
for Alabama, we have 4 obervations for series column:  Alabama, Alabama Clearances, United States, and United States Clearances. We will focus on just "Alabama", because that 
obervation represents the entirety of homicide rates per 100,000 residents in that specific month and year. We will do the same process for all the other states. 
In terms of data cleaning, tidying, and wrangling, we will first combine the 50 datasets into 1. Working with all 50 datasets will be painful, hence from each database, 
we will extract the required information and make a new database. Specifically, from each database, we will need to just extract the one observation (state name) that has the homicide rate per
100,000 by state and not state clearances, US clearances, or US. From here, we will have a "U.S .State" column for better clarity and we will also
sort the dataframe we create by the U.S. State Name. The U.S. State Name will be a categorical variable. We will make sure that we don't have any NaN values. If we have
NaN values, then we simply did not get a homicide rate per 100,000 from the FBI for that month/year. We replace that NaN value with zero. We don't have to worry 
about the time frame being incorrect because the source where we collected our data from, the FBI Crime Data Explorer, allowed us to extract data based on time requirements. From here,
we will convert the dataset from wide to long format for tidying purposes. We will split the month-year columns which are in the form of 01-1985 etc into proper
Month and Year columns, and we will map the months which are from 1-12 to Jan-Feb. This will make the dataset a little more tidy.'''


#Code For Above Data Cleaning, Tidying, And Wrangling Procedure

import glob
import os

file_paths = glob.glob("Homicide Rates Per 100,000 By US State 1985-2016/*.csv") # Use glob packaage to get the file paths.
columns_set = False # Variable For Extracting The Column Names Once
state_data = {} # Dictionary Will Store The Row We Need From Each State Dataset & Then Will Be Converted To Final DataFrame
date_lables = None # Store The Column Names

for file in file_paths :


    state_name = file.split("/")[-1].replace(".csv", "") # Get The State Name
   
    # Read CSV file
    state_df = pd.read_csv(file, header=None) # Read That State Data In
   
    # Find the row where the first column matches the state name
    state_row = state_df[state_df.iloc[:, 0] == state_name] # Extract Necessary Observation, Not Clearances.
   
    if not state_row.empty:
        # Store homicide rates for that state
        state_data[state_name] = state_row.iloc[0, 1:].values


        if not columns_set:
            date_labels = state_df.iloc[0, 1:].values.astype(str)  # Ensure date labels /column names are strings
            columns_set = True # Make sure we don't copy over the column names more than once.


# Convert dictionary to DataFrame (states as rows, dates as columns)
state_final_df = pd.DataFrame.from_dict(state_data, orient="index")
# Assign column names using the first dataset's date labels
state_final_df.columns = date_labels # Set The Columns

# Reset the index and make the US state name a regular column
state_final_df.reset_index(inplace=True)

# Rename the index column to 'US State' to make it clear that it's the state column
state_final_df.rename(columns={'index': 'U.S State'}, inplace=True)

#Ensure We Replace NaN values with zero
state_final_df.fillna(0, inplace=True)

#Sort The Database By US State (This Will Align With The Alcohol Database We Collected Above)
state_final_df = state_final_df.sort_values(by = ['U.S State'])
state_final_df = state_final_df.reset_index()
state_final_df = state_final_df.drop('index', axis = 1)

# Use melt to convert the data from wide to long format (Start Tidying)
state_final_df = state_final_df.melt(id_vars=["U.S State"], var_name="MonthYear", value_name="Homicide Rate Per 100,000 Residents")

# Now, split the 'MonthYear' column (e.g., "01-1985") into separate 'Month' and 'Year' columns.
state_final_df[['Month', 'Year']] = state_final_df['MonthYear'].str.split('-', expand=True)

# Optionally convert Month and Year to integers.
state_final_df['Month'] = state_final_df['Month'].astype(int)
state_final_df['Year'] = state_final_df['Year'].astype(int)

# Drop the now-unnecessary 'MonthYear' column.
state_final_df.drop(columns=["MonthYear"], inplace=True)

# Reorder columns for clarity.
state_final_df = state_final_df[['U.S State', 'Year', 'Month', 'Homicide Rate Per 100,000 Residents']]

month_map = {1: "January", 2: "February", 3: "March", 4: "April", 
             5: "May", 6: "June", 7: "July", 8: "August", 
             9: "September", 10: "October", 11: "November", 12: "December"}

state_final_df['Month'] = state_final_df['Month'].map(month_map)

#Test Statements To Ensure No Nan Values, & We Have Correct Shape
#Assert that we have no nan values 
null_rows2 = len(state_final_df) - len(state_final_df.dropna()) 
assert null_rows2 == 0

#Correct Shape
assert state_final_df.shape == (19200, 4)

print(state_final_df.head(100)) # See First 100
print(state_final_df.tail(100)) # See Last 100
```

### Section 3 of EDA - EDA Of Avg Homicide Rate Per 100,000 Residents & Per Capita Alcohol Consumption Combined

In this 3rd section of our EDA, we will first merge the data from our alcohol consumption dataset and the average homicide rate per 100,000 residents (mean which was calculated from the 12 months of data for each year). From there, we will look at some initial descriptive statistics about each column in the merged dataset. Then, we will make a scatterplot for evaluating the correlation between the All Beverages (Per Capita Consumption In Litres) and Average Homicide Rate Per 100,000 Residents across all 50 states in every 10 years from 1985 to 2016. Evaluating the pearson coefficients from 1985 to 2016 in 10 year intervals can help us understand how the correlation might have evolved over the years, allowing us to perform a level of time analysis. We only perform in 10 year intervals due to limitations of rendering of plots on GitHub. We use Seaborn and Matplotlib to help generate the scatterplots

```python

# First, we merge the data from the alcohol consumption dataset and the average homicide rate for each state for each year.
merged_alcohol_homicide_df = pd.merge(alcohol_df, state_final_df_agg, on=['U.S State', 'Year'])
# Look at the data
merged_alcohol_homicide_df.head()
```

```python
# From there, we can take a look at some descriptive statistics regarding the data.

#Look At Descriptive Statistics For Beer Consumption For All 50 States From 1985-2016
summary_beer = merged_alcohol_homicide_df["Beer (Per Capita Consumption In Litres)"].describe()
print(summary_beer)

#Look At Descriptive Statistics For Wine Consumption For All 50 States From 1985-2016
summary_wine = merged_alcohol_homicide_df["Wine (Per Capita Consumption In Litres)"].describe()
print(summary_wine)
```

```python
#Look At Descriptive Statistics For Spirits Consumption For All 50 States From 1985-2016
summary_spirits = merged_alcohol_homicide_df["Spirits (Per Capita Consumption In Litres)"].describe()
print(summary_spirits)

#Look At Descriptive Statistics For All Beverages Consumption For All 50 States From 1985-201
summary_alcohol = merged_alcohol_homicide_df["All Beverages (Per Capita Consumption In Litres)"].describe()
print(summary_alcohol)
```

```python
#Look At Descriptive Statistics For Average Homicide Rate Per 100,000 Residents For All 50 States From 1985-2016
summary_homicide = merged_alcohol_homicide_df["Average Homicide Rate Per 100,000 Residents"].describe()
print(summary_homicide)
```

```python

# Create a list of years from 1985 to 2016 (incrementing by 10)
years = list(range(1985, 2017, 10))

# Set Seaborn style
sns.set(style="whitegrid")

# Create plots for each year
for year in years:
    # Filter data for the current year
    year_data = merged_alcohol_homicide_df[merged_alcohol_homicide_df['Year'] == year]
    
    if year_data.empty:
        print(f"No data available for {year}")
        continue

    # Compute Pearson correlation coefficient
    correlation = np.corrcoef(
        year_data["All Beverages (Per Capita Consumption In Litres)"], 
        year_data["Average Homicide Rate Per 100,000 Residents"]
    )[0, 1]
    
    # Create the scatter plot with a larger figure size
    plt.figure(figsize=(12, 7))  # Bigger plot
    sns.scatterplot(
        data=year_data, 
        x="All Beverages (Per Capita Consumption In Litres)", 
        y="Average Homicide Rate Per 100,000 Residents", 
        hue="U.S State",  # Color points by state
        palette="viridis", 
        edgecolor="black"
    )
    
    # Add title with correlation
    plt.title(f"{year}: Per Capita Alcohol Consumption (All Beverages) vs Avg Homicide Rate Per 100,000 Residents\nPearson Coefficient: {correlation:.2f}", fontsize=14)
    plt.xlabel("Alcohol Consumption (Litres Per Capita, All Beverages)", fontsize=12)
    plt.ylabel("Average Homicide Rate (Per 100K Residents)", fontsize=12)

    # Adjust legend: make it horizontal at the bottom
    plt.legend(
        title="U.S State", 
        loc="upper center", 
        bbox_to_anchor=(0.5, -0.12),  # Moves legend below the plot
        ncol=6,  # More columns to spread the legend
        prop={'size': 6}  # Slightly larger legend font
    )
    
    # Adjust layout to prevent cutting off the legend
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

### Section 3 Of EDA - Analysis

Looking at the scatterplots through the years, we see that as we approach 2015, we have a pearson coefficient that becomes slightly more stronger in a negative correlation, yet overall the relationship between average homicide rate per 100,000 and per capita alcohol consumption in litres (All Beverages) can be described as a weak negative correlation between the two. Our findings in this EDA suggest that there is not a very strong linear relationship between the two and that there could be several other confounding variables impacting the relationship. When considering that both average homicide rates per 100,000 and per capita alcohol consumption in litres are both topics of social science studies, our result of a weak negative correlation and a weak linear relationship makes sense as social science research and data science questions related to social science often have multifaceted influences. As far as predictive models are concerned, linear regression would not be appropriate here because we have a very weak linear relationship and there could be the potential for high autocorrelation given the time element to our EDA. 

To further substantiate the weak negative correlation, below we have a scatterplot that involves all the datapoints from all 50 states from 1985 to 2016 and we can analyze the pearson coefficient when looking at all the datapoints at once. Additionally, we also have a correlation matrix that gives us the pearson coefficients across all the variables in the dataset. Our analysis primarily focused on all alcoholic beverages, but the correlation matrix will help identify the relationships between the average homicide rate per 100,000 residents and specific types of alcohol.

```python

# Calculate Pearson correlation coefficient
correlation = np.corrcoef(
    merged_alcohol_homicide_df["All Beverages (Per Capita Consumption In Litres)"], 
    merged_alcohol_homicide_df["Average Homicide Rate Per 100,000 Residents"]
)[0, 1]

# Set figure size
plt.figure(figsize=(10, 6))

# Create scatter plot
sns.scatterplot(
    data=merged_alcohol_homicide_df,
    x="All Beverages (Per Capita Consumption In Litres)",
    y="Average Homicide Rate Per 100,000 Residents",
    alpha=0.5  # Reduce marker opacity
)

# Set labels and title
plt.xlabel("Alcohol Consumption (All Beverages Litres Per Capita)")
plt.ylabel("Average Homicide Rate (Per 100K Residents)")
plt.title(f"Scatter Plot: Alcohol Consumption (All Beverages) vs Avg Homicide Rat Per 100,000 Residents (1985-2016)\nPearson Coefficient: {correlation:.2f}")

# Show plot
plt.show()


# Compute correlation matrix

# Select only quantitative (numerical) columns
quantitative_cols = merged_alcohol_homicide_df.select_dtypes(include=['number'])

# Compute correlation matrix for numerical data
corr_matrix = quantitative_cols.corr()

# Set figure size
plt.figure(figsize=(5, 5))

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Set title
plt.title("Correlation Matrix of Quantitative Variables")

# Show plot
plt.show()

```
