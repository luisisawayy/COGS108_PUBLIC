# Research Question

-  COGS 108 Project Research Question: **What is the relationship between per capita alcohol consumption (liters of alcoholic beverages consumed per person) and homicide rates per 100,000 people across all 50 states of the US from 1985-2016 ?**.
-  Key Variables: Per Capita Alcohol Consumption, Homicide Rate, Time Frame: 1985-2016, Consideration For All 50 States

## Background and Prior Work

Alcohol consumption has often been viewed as a potential factor when considering violent crime rates in America, specifically homicides. Our topic revolves around the relationship between per capita alcohol intake and homicide rate per 100,000 residents across US states from the years 1985-2016. By performing data analysis over multiple decades, our goal through this project is to understand possible correlations between alcoholic consumption and homicide rates over the years, accounting for variations across the 50 states of America. Understanding this relationship could help provide more insight into the role of alcohol in crime dynamics and give a more detailed analysis of whether risky behaviors such as alcohol consumption and homicides rates are linked together or if they have some form of positive association. Throughout this project, we will use exploratory data analysis, inference, and regression models to help us come to strong conclusions that can provide an answer to our research question in this project. In the following paragraphs, we will explain some of the prior work already done related to alcoholism and homicide rates and the relationship between the two.

According to the National Institute Of Health, alcoholic consumption and binge drinking have increased significantly over the last 10-15 years since the early 2000s. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) According to their research, the annual increase in alcoholic consumption has been about 0.30%. NIH have also suggested an increase in alcohol related mortality rates. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The increase in alcohol consumption lays the platform perfectly for us to conduct research on it's potential association and impact on homicide rates. While NIH's research only studies the increase in alcohol consumption since the 2000s, <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) it provides some background for us to work with and will allow us to make early predictions as we work on our project. Our project will encompass a much larger time frame (1985-2016) and give a larger overview of alcoholism trends.

While alcohol consumption is a crucial factor in our study, understanding homicide rate trends is equally important. We also need to focus on homicide rates per 100,000 residents. According to the Council On Criminal Justice, homicide rates have fluctutated since the 1980s. While there was a decline in the early 1980s, there was a sharp increase in the early 1990s, followed by a steady decline until the early 2000s, after which the homicide rates peaked again in the 2010s. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Homicide rates have fluctutated a lot, suggesting that homicide rates have a complex set of dynamics driving the changes <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Our project will try to understand the role alcohol has in the changes of homicide rates. 

Beyond just understanding the changes of rates, prior research regarding the relationship between alcohol policies and violent crime provide a more solid base for us to work on. According to Christopher Carpenter and Carlos Dobkin, professors at UCI and UCSC, the sudden changes of national alcohol sale policies historically have resulted in significant increases in violent crime, especially homicide. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Their research suggests potential relationships between alcohol consumption and homicide rates because policies that would have increased alcohol access could have potentially increased consumption. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Carpenter and Dobkin's research aligns with our project’s goal of examining the long-term relationship between alcohol consumption and homicide rates across different states. While their research focused on policy changes, our project will aim to expand on their research by exploring trends and relationships over multiple decades (1985-2016) and analyzing state-by-state variations, and try to discover some of the relationships between alcohol consumption and homicide rates.

It is important to note that the homicide and alcoholism rates and the relationship discussed by the Council on Criminal Justice, NIH and the prior research we explored considered the United States as a whole, providing an overview of trends and the potential relationships at the national level. However, our project will focus on the specifics of each individual state, providing a more focused approach which will allow us to build on all the prior work done by evaluating the relationship between per capita alcoholic consumption and homicide rates per 100,000 residents on a much more granular level. Our project will allow us to analyze and compare the differences in the relationship from state to state. Analyzing from state to state opens up the possibility for us to answer more follow up questions. For example, after our research is completed, we can also answer the question of what kind of geographical patterns do we see in the relationship between per capita alcohol consumption and homicide rates per 100,000 residents ? Is there a stronger or weaker association in certain states ?

# Data

The ideal dataset for our research would capture both alcohol consumption and homicide rates across all U.S. states from 1985 to 2016. The dataset should include the following:
Variables:
State: The U.S. state the data corresponds to.
Year: The year the data was collected (1985-2016).
Per Capita Alcohol Consumption: Measured in Liters per person per year. If possible, broken down by type of alcohol (I.E. Beer, Wine, Vodka, etc.)
Homicide Rate: The number of homicides per 100,000 residents per year.
Population: Total state population to adjust for per capita analysis.
Additional Factors (if dataset allows for): Variables like median household income, unemployment rate, or gun ownership rate could serve as control variables for our research if they are available.
Number of Observations:
Given that we are covering 50 states over a 31-year period (1985 - 2016), the dataset should have 1,550 observations (50 x 31 = 1550) if the data is available for every year.
Data Collection & Organization:
Alcohol Consumption Data: Typically collected by health organizations, government agencies, and industry reports using sales data, tax records, and surveys. Accessible through the internet and database sites such as Kaggle.
Homicide Rate Data: Usually gathered from law enforcement records, FBI’s Uniform Crime Reporting (UCR) program, or the CDC’s violent death data. Also accessible through the internet and database sites such as Kaggle.
Storage Format: The data would be collected, and structured in a tidy format and be stored in CSV, SQL database, or a Pandas DataFrame, where each row represents a state-year observation, and each column corresponds to one of the key variables.
Here are some promising sources that are likely to meet our criteria for the ideal data we need for our analysis:
